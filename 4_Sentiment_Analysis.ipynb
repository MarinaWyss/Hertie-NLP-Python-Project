Import packages and import data

```python
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize import wordpunct_tokenize
from nltk.corpus import stopwords
```

```python
data = pd.read_csv('data/sentiment_test.csv')
```

### 1. Pre-processing

Tokenization

```python
# setup
documents = data['article_text']

sentence_level = []
bag_of_words = []
sent_tok_level = []
```

```python
# tokenize words at the sentence level
for item in documents:
    item = [word_tokenize(t) for t in sent_tokenize(item)]
    sentence_level.append(item)
```

```python
# tokenize bag of words
for item in documents:
    item = wordpunct_tokenize(item)
    bag_of_words.append(item)
```

Remove punctuation

Remove stop words

```python
# setup
stop_words =  set(stopwords.words('english'))

sent_no_stop = []
bag_no_stop = []
```

```python
# remove stop words
for item in sentence_level:
    item = [w for w in item if not w in stop_words]
    sent_no_stop.append(item)
    
print(sent_no_stop)
```

```python
# remove stop words
for item in bag_of_words:
    item = [w for w in item if not w in stop_words]
    bag_no_stop.append(item)
```
